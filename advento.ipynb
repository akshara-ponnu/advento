```python
#day3 part1
def cal_priority(char):
    #  lowercase item  priorities 1 through 26
    if 'a' <= char <= 'z':
        print(ord(char) - ord('a')+1)
        return ord(char) - ord('a') + 1
        
        
           
    # uppercase item  priorities 27 through 52
    elif 'A' <= char <= 'Z':
        print(ord(char) - ord('A')+27 )
        return ord(char) - ord('A') + 27
    else:
        return 0  # Return 0 for non-alphabetic characters

def find_priority_sum(rucksack):
    half_length = len(rucksack) // 2
    #print(half_length)
    first_comp = rucksack[:half_length]
    #print(first_comp)
    
    second_comp = rucksack[half_length:]
    #print(second_comp)
    common = set(first_comp) & set(second_comp)
    print(common)
    priority_sum = sum(cal_priority(item) for item in common)
    return priority_sum

# Example list of contents from six rucksacks
file_path = r"C:\Users\AKSHARA\Downloads\in.txt"
try:
    with open(file_path, 'r') as file:
        file_contents = file.read()
        #print(file_contents)
        # Split the contents into individual rucksacks
        rucksacks = file_contents.strip().split('\n')
        #print(rucksacks)

except FileNotFoundError:
    print(f"File not found at path: {file_path}")
except Exception as e:
    print(f"An error occurred: {e}") 

# Calculate and print the sum of priorities for all rucksacks
total_priority_sum = sum(find_priority_sum(rucksack) for rucksack in rucksacks)
print(f"Total Priority Sum: {total_priority_sum}")

```
